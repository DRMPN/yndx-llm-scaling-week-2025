```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

# Загрузка данных
train = pd.read_csv('/kaggle/input/llm-scaling-week/train_weights.csv')
test = pd.read_csv('/kaggle/input/llm-scaling-week/test_weights.csv')

# Подготовка признаков
X_train = train[['W0', 'W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8', 'W9']]
y_train = train['MSE']

# Создание полиномиальных признаков 2-й степени
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)

# Обучение Ridge-регрессии
model = Ridge(alpha=1.0)
model.fit(X_train_poly, y_train)

# Подготовка тестовых данных
X_test = test[['W0', 'W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8', 'W9']]
X_test_poly = poly.transform(X_test)

# Предсказание
y_pred = model.predict(X_test_poly)

# Формирование ответа
test['MSE'] = y_pred
answers = test.to_dict(orient='records')

# Сохранение в файл answers
import json
with open('answers.json', 'w') as f:
    json.dump(answers, f)
```
